### 12. Analyse visually and write your inferences about the relationship between the following:


##### i) Total amount spent Vs Dependents. (Dependents=['Kidhome']+['Teenhome'])

```python
df[['total_amount','Kidhome','Teenhome']]
```

```python
dependent = df['Kidhome'] + df['Teenhome']
dependents = dependent.to_frame(name = 'Dependents')
dependents
```

```python
spent_df = pd.concat([df['total_amount'],dependents],axis = 1)
spent_df
```

```python
dep_group = spent_df.groupby(by = 'Dependents')['total_amount']
total_sum = dep_group.sum()
dep_df = total_sum.to_frame(name = 'Total')
dep_df
```

```python
#Plotting Bar plot to compare dependants and no of responses
ax = dep_df.plot(kind = 'bar')
ax.set_ylabel('Number of responses',fontsize = 12)
plt.show()
```

##### ii) Total Purchases Vs Dependents

```python
#finding Total purchases wrt Dependants
purchases_df = pd.concat([df['total_purchase'],dependents],axis = 1)
purchases_df
```

```python
purchase_group = purchases_df.groupby(by = 'Dependents')['total_purchase']
purchase_df = purchase_group.sum().to_frame(name = 'Nopurchases')
purchase_df
```

```python
#Plotting Bar Plot to compare Dependants and no of purchases
ax = purchase_df.plot(kind='bar')
ax.set_xlabel('Dependents',fontsize = 12)
ax.set_ylabel('Num of purchases',fontsize = 12)
plt.show()
```
